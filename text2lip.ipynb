```
#@title AI Text summary to lip sync avatar started! Please upload your text!
#@markdown 1. Enter Text To Summarize (up to 500 words):
input_text = 'What is multilayer perceptron?' #@param {type:"string"}
#@markdown 2. Choose speaker voice
speaker_voice = 'default' #@param ['Robert', 'Sam', 'Anatoly'] {allow-input: true}
#@markdown 3. Choose Video on which creaate the lip sync (up to 20 seconds):
face_video = "hitla.mp4" #@param ['indian.mp4', 'hitla.mp4'] {allow-input: true}
#@markdown 4. Press the play (triangle) button on the left.

import subprocess
import os

prompt = f"""Instruct: Answer the given question:{input_text}\nOutput:"""
inputs = tokenizer(prompt, return_tensors="pt", return_attention_mask=False)
outputs = model.generate(**inputs, max_length=256)
text = tokenizer.batch_decode(outputs)[0]
cleaned_text = text.split("Output: ",1)[1]
cleaned_text = cleaned_text.split("<|endoftext|>")[0]

print("Model Response Text")
print(cleaned_text)

output_path = "/content/data/output_beta.wav"
command = f'tts --text "{cleaned_text}" --out_path {output_path}'

try:
    print("Converting text to audio...")
    subprocess.run(command, shell=True, check=True)
    print("Text-to-speech conversion successful!")
except subprocess.CalledProcessError as e:
    raise Exception(f"Text to audio converted with error: {e}")

full_face_path = os.path.join("/content/gdrive/MyDrive/sample_audio/", face_video)
full_audio_path = '/content/data/output_beta.wav'
full_outfile_path = '/content/data/result.mp4'
checkpoint_path = 'checkpoints/wav2lip_gan.pth'

print("Creating lip sync video from audio...")
command_wav2lip = f'cd Wav2Lip && python inference.py --checkpoint_path {checkpoint_path} --face "{full_face_path}" --audio "{full_audio_path}" --resize_factor 2 --outfile "{full_outfile_path}"'


try:
    print("Converting audio to lip sync...")
    subprocess.run(command_wav2lip, shell=True, check=True)
    print("Audio to speach conversion successful!")
except subprocess.CalledProcessError as e:
    raise Exception(f"Audio to speach converted with error: {e}")

print("FINISHED!!!")


from IPython.display import HTML
from base64 import b64encode

def show_video(video_path, video_width = 600):

  video_file = open(video_path, "r+b").read()

  video_url = f"data:video/mp4;base64,{b64encode(video_file).decode()}"
  return HTML(f"""<video width={video_width} controls><source src="{video_url}"></video>""")

show_video("data/result.mp4")
```
